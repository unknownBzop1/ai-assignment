```python
import numpy as np
import pandas as pd
import os
```

```python
path = 'C:/PycharmProjects/revamped/ai-assignment'

image_labels_df = pd.read_csv(os.path.join(path, 'Data_Entry_2017_v2020.csv'))
image_labels_df
```

```python
image_labels_df.query('`Finding Labels` == \'No Finding\'')
```

```python
image_labels_df['label_set'] = image_labels_df['Finding Labels'].apply(lambda labels: set(labels.split('|')))
```

```python
from functools import reduce

labels: set = reduce(lambda x, y: x.union(y), image_labels_df['label_set'])
```

```python
for label in labels:
    print(label, image_labels_df[image_labels_df['label_set'] == {label}].shape)
```

```python
pneumothorax_df = image_labels_df.query('`Finding Labels` == \'Pneumothorax\'')
pneumothorax_df
```

```python
image_labels_df.query('`Finding Labels` == \'Pneumothorax\'').groupby(['Patient Sex', 'View Position']).agg({'Image Index': 'count'}).reset_index()
```

```python
f_ap_count = len(image_labels_df.query('`Finding Labels` == \'Pneumothorax\' and `Patient Sex` == \'F\' and `View Position` == \'AP\'')['Image Index'].unique())
f_pa_count = len(image_labels_df.query('`Finding Labels` == \'Pneumothorax\' and `Patient Sex` == \'F\' and `View Position` == \'PA\'')['Image Index'].unique())
m_ap_count = len(image_labels_df.query('`Finding Labels` == \'Pneumothorax\' and `Patient Sex` == \'M\' and `View Position` == \'AP\'')['Image Index'].unique())
m_pa_count = len(image_labels_df.query('`Finding Labels` == \'Pneumothorax\' and `Patient Sex` == \'M\' and `View Position` == \'PA\'')['Image Index'].unique())
```

```python
no_finding_df = image_labels_df.query('`Finding Labels` == \'No Finding\'')
no_finding_f_ap_samples = no_finding_df.query('`Patient Sex` == \'F\' and `View Position` == \'AP\'').sample(n=f_ap_count, random_state=0, replace=False)
no_finding_f_pa_samples = no_finding_df.query('`Patient Sex` == \'F\' and `View Position` == \'PA\'').sample(n=f_pa_count, random_state=0, replace=False)
no_finding_m_ap_samples = no_finding_df.query('`Patient Sex` == \'M\' and `View Position` == \'AP\'').sample(n=m_ap_count, random_state=0, replace=False)
no_finding_m_pa_samples = no_finding_df.query('`Patient Sex` == \'M\' and `View Position` == \'PA\'').sample(n=m_pa_count, random_state=0, replace=False)
```

```python
no_finding_samples = pd.concat([no_finding_f_ap_samples, no_finding_f_pa_samples, no_finding_m_ap_samples, no_finding_m_pa_samples])
no_finding_samples
```

```python
training_test_df = pd.concat([no_finding_samples, pneumothorax_df])
training_test_df
```
